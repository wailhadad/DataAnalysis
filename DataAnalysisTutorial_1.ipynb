```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
path_to_csv='/content/drive/MyDrive/Colab Notebooks/sales_data.csv'
df=pd.read_csv(path_to_csv, parse_dates=['Date'])
df.head()
```

```python
df['Age_Group'].value_counts().plot(
    kind='pie', subplots=True,
    title='Pie Chart',
    autopct='%1.2f%%',
    figsize=(6,6),
    textprops={'fontsize': 10},
    ylabel='proportion',
)
```
